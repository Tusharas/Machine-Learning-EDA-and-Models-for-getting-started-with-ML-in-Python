```python
#5.For data_bening.radius_mean column find the outliers and print the range of outliers and outliers values.
#The interquartile range (IQR), also called the midspread or middle 50%, or technically H-spread, is a measure of statistical dispersion, being equal to the difference between 75th and 25th percentiles, or between upper and lower quartiles, IQR = Q3 âˆ’ Q1.
data_bening_radius_mean=  data3.loc[data3['diagnosis'] == 'B', 'radius_mean']

Q1 = np.percentile(data_bening_radius_mean, 25, interpolation = 'midpoint')  
Q2 = np.percentile(data_bening_radius_mean, 50, interpolation = 'midpoint')  
Q3 = np.percentile(data_bening_radius_mean, 75, interpolation = 'midpoint')  
  
print('Q1 25 percentile of the given data is, ', Q1) 
print('Q1 50 percentile of the given data is, ', Q2) 
print('Q1 75 percentile of the given data is, ', Q3) 
  
IQR = Q3 - Q1  
print('Range is', IQR) 
print(data3['radius_mean'] < (Q1 - 1.5 * IQR)) , (data3['radius_mean'] > (Q3 + 1.5 * IQR))
```

```python
#7.Find the correlation between all features using Correlation Map.
corrMatrix = data3.corr()
plt.figure(figsize=(10,15))
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
#In the covariance matrix in the output, the off-diagonal elements contain the covariances of each pair of variables. 
#The diagonal elements of the covariance matrix contain the variances of each variable.
#The variance measures how much the data are scattered about the mean.
data3.cov()
```
